```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('CASE3.1_real.csv')

# Calculate min and max for each column
min_vals = df.min()
max_vals = df.max()

# Iterate through columns and print the formatted min/max tuple
for col in df.columns:
    # Check if the column is numeric to apply rounding
    if pd.api.types.is_numeric_dtype(df[col]):
        min_val = round(min_vals[col], 3)
        max_val = round(max_vals[col], 3)
        print(f"{col} ({min_val}, {max_val})")
    else:
        # For non-numeric columns, just print the values as is
        print(f"{col} ({min_vals[col]}, {max_vals[col]})")
```
